Здесь соберем всю обработку для усорения подгрузки данных. Пользоваться будем `pipe` методами `Pandas`. В процссе изучения данных, было обнаружено, что по ИП отсутствует около 70% информации, что присуща Юридическим лицам. Было принято решение строить две модели - на ИП и на ЮЛ

````python
class ShapFeatureSelection (BaseEstimator, TransformerMixin):
  """
  Класс вычисляет  важность признаков на основе библиотеки Shap. Работает 
  только с деревьями для большей эффективности или моделями на основе 
  градиентного бустинга. Приоритетно использовать такие модели как:
   
   Catboostclassifier, XGboostclassifier

  Для работы необходимо импортировать:

    from sklearn.base import BaseEstimator, TransformerMixin
    import shap

  Параметры
    ----------
    estimator : 
        Обучение с учителем с методом fit позволит получить и отобрать индексы
        самых важных признаков
    n_features_to_select : int, default=None
        Количество признаков для отбора, по умолчанию значение None
    columns: List, default=None
        Список признаков исходного сета, по умолчанию значение None
    
  Методы
    ----------
    fit - обучается и выявляет наиболее важные признаки
    tranform - изменяет исходный сет и вовзращает отобранные признаки
    get_index - возвращает отобранные признаки индексов
    plot_values - построение графика shap values
    _estimator_type - метод @property - возвращает тип модели
    get_feature_importance - возвращает DataFrame FI
  Примечание
    -----
    Не работает с категориальными признаками. Nan/Inf разрешены в случае, если 
    их принимает модель метода fit

    """

  def __init__(self, estimator, n_features_to_select=None, columns=None):
    self.estimator = estimator
    self.n_features_to_select = n_features_to_select
    self.columns = columns

  def fit(self, X, y, **fit_params):
    # разделить выборку на обучение и валидацию
    X_train, self.X_val, y_train, y_val = train_test_split(
                                      X, y, test_size=.25, random_state=42)
    
    assert X_train.shape[0] + self.X_val.shape[0] == X.shape[0], 'Неправильно разделена выборка'
    # обучить модель
    self.estimator.fit(X_train, y_train, eval_set = [(self.X_val, y_val)],
                       early_stopping_rounds = 200
                       , use_best_model=True)


    # получаем предсказание веростности классов
    preds = self.estimator.predict_proba(self.X_val)[:, 1]
    explainer = shap.TreeExplainer(self.estimator)
    self.shap_values = explainer.shap_values(self.X_val)
    # собираем сет данных
    try:
    
      vals = np.abs(self.shap_values).mean(0)
      self.feature_importance = pd.DataFrame(list(zip(self.columns, vals)),columns=['col_name','feature_importance_vals'])
      self.feature_importance.sort_values(by=['feature_importance_vals'],ascending=False,inplace=True)
      self.idx = list(self.feature_importance['feature_importance_vals'].index)[:self.n_features_to_select]
      
    except:
      
      vals = np.abs(self.shap_values[:1]).mean(0)
      self.feature_importance = pd.DataFrame(list(zip(self.columns, vals)),columns=['col_name','feature_importance_vals'])
      self.feature_importance.sort_values(by=['feature_importance_vals'],ascending=False,inplace=True)
      self.idx = list(self.feature_importance['feature_importance_vals'].index)[:self.n_features_to_select]

    return self

  def transform(self, X, y=None):
    X_selected = X.iloc[:, self.idx]
    return X_selected

  def get_index(self):
    return self.idx

  def plot_values(self):
    return shap.summary_plot(self.shap_values, self.X_val)

  @property
  def _estimator_type(self):
    return self.estimator._estimator_type

  def get_feature_importance(self):
    return self.feature_importance
````

````python
def time_to_time(data):

  '''
  функция для обработки дат

  '''
  cols = ['SIGN_DATE', 'DATEFIRSTREG', 'TAXREG_REGDATE', 'TAXREGPAY_REGDATE']
  for col in cols:
      data[col] = data[col].fillna(data['SIGN_DATE'])


  for col in cols:
      data[col] = data[col].apply(lambda x: datetime.strptime(x, '%d%b%Y:%H:%M:%S').date())
  return data


def create_date_features(data):
  '''
  функция создает дополнительные признаки по датам. основная - время жизни контракт
  сами даты удаляются
  '''
  data = data.assign(
      
      current = datetime.now().date(),
      total_lifetime = lambda x: ((x['current'] - x['SIGN_DATE']) 
                                  / np.timedelta64(1, 'M')
                                  ),
      lifetime = lambda x: ((x['SIGN_DATE'] - x['DATEFIRSTREG']) 
                                  / np.timedelta64(1, 'M')
                                  ),
      month_for_reg = lambda x: ((x['TAXREG_REGDATE'] - x['DATEFIRSTREG']) 
                                  / np.timedelta64(1, 'M')
                                  ),
      month_from_payreg = lambda x: ((x['TAXREG_REGDATE'] - x['TAXREGPAY_REGDATE']) 
                                  / np.timedelta64(1, 'M')
                                  )
                     
      
  )
  data['lifetime'] = data['lifetime'].replace(0, data['lifetime'].mean())
  data['month_for_reg'] = data['month_for_reg'].replace(0, data['month_for_reg'].mean())
  data['month_from_payreg'] = data['month_from_payreg'].replace(0, data['month_from_payreg'].mean())
  cols = ['SIGN_DATE', 'DATEFIRSTREG', 'TAXREG_REGDATE', 'TAXREGPAY_REGDATE', 'current']
  data_new = data.drop(cols, axis=1).copy()



  return data_new


def drop_corr_current(data, lvl):
  '''
  функция удаляет признаки с сильной корреляцией по текущим доходам
  '''
  data_corr_current = data.loc[:, 'F1100':'F2410'].copy()
  phik_overview = data_corr_current.phik_matrix().abs()
  upper_tri = phik_overview.where(np.triu(np.ones(phik_overview.shape), k=1).astype(np.bool))
  to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > lvl)]
  data_new = data.drop(to_drop, axis=1)
  return data_new


def drop_corr_lag(data, lvl):
  '''
  функция удаляет признаки с сильной корреляцией по доходам за прошлый период
  '''
  data_corr_current = data.loc[:, 'F1150_LAG1':'F2400_LAG1'].copy()
  phik_overview = data_corr_current.phik_matrix().abs()
  upper_tri = phik_overview.where(np.triu(np.ones(phik_overview.shape), k=1).astype(np.bool))
  to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > lvl)]
  data_new = data.drop(to_drop, axis=1)
  return data_new


def drop_corr_court(data, lvl):
  '''
  функция удаляет признаки с сильной корреляцией по судам
  '''
  data_corr_current = data.loc[:, 'PLAINTIFF_CASESNUMBER_YEAR':'THIRDOROTHERPERSON_EVER'].copy()
  phik_overview = data_corr_current.phik_matrix().abs()
  upper_tri = phik_overview.where(np.triu(np.ones(phik_overview.shape), k=1).astype(np.bool))
  to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > lvl)]
  data_new = data.drop(to_drop, axis=1)
  return data_new


def drop_corr_tender(data, lvl):
  '''
  функция удаляет признаки с сильной корреляцией по тендерам
  '''
  data_corr_current = data.loc[:, 'ADMITTEDNUMBER_233_YEAR':'SUM_95_EVER'].copy()
  phik_overview = data_corr_current.phik_matrix().abs()
  upper_tri = phik_overview.where(np.triu(np.ones(phik_overview.shape), k=1).astype(np.bool))
  to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > lvl)]
  data_new = data.drop(to_drop, axis=1)
  return data_new

def fill(data):
  '''
  функция заполняет пропуски. Доходы не затрагиваются
  '''
  
  cols_0 = ['COUNTCOOWNERFCSM', 'COUNTCOOWNERROSSTAT', 'COUNTBRANCH',
            'COUNTBRANCHROSSTAT', 'COUNTBRANCHEGRUL', 'COUNTCOOWNEREGRUL']



  for col in cols_0:
    data[col] = data[col].fillna(0)


  cols = ['FLAG_DISQUALIFICATION', 'COUNT_CHANGE_YEAR', 'COUNT_CHANGE_EVER', 'EGRPOINCLUDED']
  for col in cols:
    data[col] = data[col].fillna(0)
    
    

  col_mode = ['SEX_NAME', 'CITIZENSHIP_NAME']
  for col in col_mode:
    data[col] = data[col].fillna('неизвестный')
  
  col_prop = ['OKFS_GROUP', 'OKOPF_GROUP', 'OKOGU_GROUP']
  for col in col_prop:
    data[col] = data[col].fillna('индивидуальный предприниматель')


  col_okato = ['OKTMO_CODE', 'OKVED_CODE']
  for col in col_okato:
    data[col] = data[col].fillna('отсутствует')

  col_mode_val = ['WORKERSRANGE', 'OKATO_REGIONCODE', 'OKATO_FED']
  for col in col_mode_val:
    data[col] = data[col].fillna(data[col].mode()[0])

  col_mean_val = ['AGE']
  for col in col_mean_val:
    data[col] = data[col].fillna(data[col].mean())

  data['EGRPOINCLUDED'] = data['EGRPOINCLUDED'].astype('int64')
  data['FLAG_DISQUALIFICATION'] = data['FLAG_DISQUALIFICATION'].astype('int64')
  data['OKATO_REGIONCODE'] = data['OKATO_REGIONCODE'].astype('int64')
  data['OKATO_FED'] = data['OKATO_FED'].astype('int64')

  return data

def create_flag_features(data):
  '''
  Создаем флаги, что данные от лица и росстата не соответствуют
  '''
  data = data.assign(
      


      count_flag = lambda x: ((x['COUNTCOOWNERFCSM']==x['COUNTCOOWNERROSSTAT'])*1),
      branch_flag = lambda x: ((x['COUNTBRANCH']==x['COUNTBRANCHROSSTAT'])*1),
      
  )

  return data

def drop_feature(data):
  '''
  удаляем столбцы без информации в обучающей выборке
  '''
  cols = ['id_contract', 'id_client', 'BIRTHDATE', 'OKTMO_CODE', 'OKTMO_FED', 'OKVED_CODE']
  for col in cols:
    data_new = data.drop(cols, axis=1).copy()

  return data_new


def category(data):
  '''
  приводим категорильные данные к category
  '''
  cols = ['EGRPOINCLUDED', 'OKFS_GROUP', 'OKOPF_GROUP', 'OKOGU_GROUP', 
          'FLAG_DISQUALIFICATION', 'SEX_NAME', 'CITIZENSHIP_NAME', 'count_flag', 'branch_flag',
          'WORKERSRANGE', 'OKTMO_CODE', 'OKTMO_FED'] 

  for col in cols:
    data[col] = data[col].astype('category')

def prepare_ip(data):
  '''
  отбираем ИП
  '''
  data_new = data.loc[data['IP_flag'] == 1].dropna(thresh=2, axis=1)
  data_new.drop(['OKFS_GROUP', 'OKOGU_GROUP', 'OKOGU_GROUP'], axis=1, inplace=True)
  return data_new


def drop_feature_test(data):
  cols = ['id_client', 'BIRTHDATE', 'OKTMO_CODE', 'OKTMO_FED', 'OKVED_CODE']
  for col in cols:
    data_new = data.drop(cols, axis=1).copy()

  return data_new

````

````python
X_train = (pd.read_csv('/content/train_dataset_hackathon_mkb.csv', encoding='cp1251', sep=';')
            .pipe(time_to_time)
            .pipe(create_date_features)
            .pipe(fill)
            .pipe(create_flag_features)
            .pipe(drop_feature)
            .pipe(drop_corr_current, lvl=.95)
            .pipe(drop_corr_lag, lvl=.95)
            .pipe(drop_corr_court, lvl=.95)
            .pipe(drop_corr_tender, lvl=.95)

)

X_train_ip = X_train.loc[X_train['IP_flag'] == 1].dropna(thresh=2, axis=1)
X_train_ul = X_train.loc[X_train['IP_flag'] == 0]

y_train_ip = X_train_ip[['TARGET']]
y_trainp_ul = X_train_ul[['TARGET']]

X_train_ip.drop(['IP_flag', 'TARGET'], inplace=True, axis=1)
X_train_ul.drop(['IP_flag', 'TARGET'], inplace=True, axis=1)

X_test = (pd.read_csv('/content/test_dataset_hackathon_mkb.csv', encoding='cp1251', sep=';', index_col='id_contract')
            .pipe(time_to_time)
            .pipe(create_date_features)
            .pipe(fill)
            .pipe(create_flag_features)
            .pipe(drop_feature_test)
)


X_test_ip = X_test.loc[X_test['IP_flag'] == 1].loc[:, list(X_train_ip.columns)]
X_test_ul = X_test.loc[X_test['IP_flag'] == 0].loc[:, list(X_train_ul.columns)]


assert X_train_ip.shape[1] == X_test_ip.shape[1], 'Проверить функции, не равны количество столбцов по ИП'
assert X_train_ul.shape[1] == X_test_ul.shape[1], 'Проверить функции, не равны количество столбцов по ЮЛ'
assert (X_test_ul.shape[0] + X_test_ip.shape[0]) == X_test.shape[0], 'Проверить функции, неверно отобрались ИП и ЮЛ в тесте'
assert (X_train_ul.shape[0] + X_train_ip.shape[0]) == X_train.shape[0], 'Проверить функции, неверно отобрались ИП и ЮЛ'
````

## Optuna ИП

````python
def objective(trial,data=X_train_ip,target=y_train_ip):
    
    X_train_cat, X_valid_cat, y_train_cat, y_valid_cat = train_test_split(X_train_ip, y_train_ip, stratify=target,
                                                                          train_size=0.7)
    
    cat_features = list(X_train_ip.select_dtypes(include=['object', 'category']).columns)
    
    params_space = {

        
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.1, 1),
        'objective': 'Logloss',
        'depth': trial.suggest_int('depth', 2, 9),
        'boosting_type': 'Ordered',
        'bootstrap_type': 'MVS',
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 50),
        'n_estimators': 10000,
        'learning_rate':trial.suggest_float('learning_rate', 0.1, 0.4)
    }


    train_data = Pool(data=X_train_cat, 
                  label=y_train_cat,
                  cat_features=cat_features)
    valid_data = Pool(data=X_valid_cat, 
                      label=y_valid_cat,
                      cat_features=cat_features)

    model = CatBoostClassifier(**params_space, random_seed=42, verbose=False)
    model.fit(train_data,
              eval_set=valid_data, 
              use_best_model=True,
              early_stopping_rounds=200
             )
    preds = model.predict_proba(X_valid_cat)[:, 1]
    auc = roc_auc_score(y_valid_cat, preds, average ='macro')
    return auc
````

## Optuna ЮЛ

````python
def objective(trial,data=X_train_ul,target=y_trainp_ul):
    
    X_train_cat, X_valid_cat, y_train_cat, y_valid_cat = train_test_split(X_train_ul, y_trainp_ul, stratify=target,
                                                                          train_size=0.7)
    
    cat_features = list(X_train_ul.select_dtypes(include=['object', 'category']).columns)
    
    params_space = {

        
        'colsample_bylevel': trial.suggest_float('colsample_bylevel', 0.1, 1),
        'objective': 'Logloss',
        'depth': trial.suggest_int('depth', 2, 9),
        'boosting_type': 'Ordered',
        'bootstrap_type': 'MVS',
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 50),
        'n_estimators': 10000,
        'learning_rate':trial.suggest_float('learning_rate', 0.1, 0.4)
    }


    train_data = Pool(data=X_train_cat, 
                  label=y_train_cat,
                  cat_features=cat_features)
    valid_data = Pool(data=X_valid_cat, 
                      label=y_valid_cat,
                      cat_features=cat_features)

    model = CatBoostClassifier(**params_space, random_seed=42, verbose=False)
    model.fit(train_data,
              eval_set=valid_data, 
              use_best_model=True,
              early_stopping_rounds=200
             )
    preds = model.predict_proba(X_valid_cat)[:, 1]
    auc = roc_auc_score(y_valid_cat, preds, average ='macro')
    return auc
````
